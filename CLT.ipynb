# The Central Limit Theorem

## Theorem:
Assume i.i.d. $X_1, X_2, ..., X_n, ...$ with finite mean $\mu$ and finite variance $\sigma^2$. As $n\rightarrow\infty$
$$\sqrt{n}\left(\frac{\overline{X}_n - \mu}{\sigma}\right) \rightarrow \mathcal{N}(0, 1)$$

${\textbf{Meaning}}$ : the CDF of l.h.s. converges to the CDF of the standart Normal distribution.

## Task:

Repeat the following m times:
- Pick any distribution
- Repeat the following $m$ times:
    * Draw $n$ samples from it
    * Sum the samples, subtract mean, divide by variance
- Plot the histogram on m values and standard normal PDF on one plot


## Discrete distribution: Poisson
$$f(k;\lambda )=P(X{=}k)={\frac {\lambda ^{k}e^{-\lambda }}{k!}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sts
import random


n = 100000
lambd = 50

# generate random variable with poisson distribution
s = np.random.poisson(lambd, n)

means = []
num_samples = [100, 5000, 10000, 50000]
n = 500
for i in num_samples:
    mean_per_sample = []
    for m in range(0, i):
        sample = random.choices(s, k=n)
        mean_per_sample.append(sum(sample)/len(sample))
    mu = np.mean(mean_per_sample)
    std = np.std(mean_per_sample)
    z = (mean_per_sample - mu) / std
    means.append(z)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 10))
sns.set_theme(style="darkgrid")
mu = 0
sigma = 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = sts.norm.pdf(x, mu, sigma)
point = dict(zip(x, y))
k = 0
for i in range(0, 2):
    for j in range(0, 2):
        sns.histplot(ax=axes[i, j], data=means[k], stat='density')
        axes[i, j].plot(x, y, color='red')
        axes[i, j].set_title(label=f"Number of samples: {num_samples[k]}")
        k += 1
plt.show()
```

## Continuous distribution: Exponential

$$f(x) = \begin{cases} \lambda e^{-\lambda x}, \; x \geq 0 \\ 0, \; x < 0 \end{cases}$$

```python
n = 100000
rate = 50

# generate random variable with exponential distribution
s = np.random.exponential(rate, n)

means = []
num_samples = [100, 5000, 10000, 50000]
n = 500
for i in num_samples:
    mean_per_sample = []
    for m in range(0, i):
        sample = random.choices(s, k=n)
        mean_per_sample.append(sum(sample)/len(sample))
    mu = np.mean(mean_per_sample)
    std = np.std(mean_per_sample)
    z = (mean_per_sample - mu) / std
    means.append(z)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 10))
sns.set_theme(style="darkgrid")
mu = 0
sigma = 1
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = sts.norm.pdf(x, mu, sigma)
point = dict(zip(x, y))
k = 0
for i in range(0, 2):
    for j in range(0, 2):
        sns.histplot(ax=axes[i, j], data=means[k], stat='density')
        axes[i, j].plot(x, y, color='red')
        axes[i, j].set_title(label=f"number of samples: {num_samples[k]}")
        k += 1
plt.show()
```
